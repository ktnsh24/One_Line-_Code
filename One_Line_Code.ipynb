```python
# Provides a DataFrame structure to store data in memory and work with it easily and efficiently.
import pandas as pd
#Provides a fast numerical array structure and helper functions.
import numpy as np


#Python libraries for the Visualization of data
#Basic plotting library in Python; most other Python plotting libraries are built on top of it.
import matplotlib.pyplot as plt
#Advanced statistical plotting library.
import seaborn as sb
# This line tells the notebook to show plots inside of the notebook
%matplotlib inline


import sys
from IPython.display import display, HTML

formatters = {             
    'RED': '\033[91m',     
    'GREEN': '\033[92m',   
    'END': '\033[0m',      
}
dataset = dataset
d = {'' : [len(dataset.columns),len(dataset.index)]}
Dataset_info =pd.DataFrame(d, index=['Number_of_variables', 'Number_of_observations'])
print '\033[1m' +"Dataset_info:"
display(Dataset_info)

print('_'*50)

print "\033[1m" + "Shape of the data:"

def shape(dataset_shape):
    Shape = dataset_shape
    return (Shape)

dataset_shape=dataset.shape
Shape = shape(dataset_shape)
print("Input data has {RED}{} observations{END} and {GREEN}{} variables{END}".format(len(dataset), len(dataset.columns),**formatters))



print('_'*50)

print '\033[1m' +"Top observations of the data:"

def head(dataset_head,x):
    Head = dataset_head
    return (Head)
aString = raw_input("Enter the number, to see the top observations: ")
a = int(aString)
x = a

dataset_head=dataset.head(x)
Head = head(dataset_head,x)
print "Head:"
display(Head)


print('_'*50)
print '\033[1m'+"Bottom observations of the data"  
def tail(dataset_tail,y):
    Tail = dataset_tail
    return (Tail)
bString = raw_input("Enter the number, to see of tail observations: ")
b = int(bString)
y = b

dataset_tail=dataset.tail(y)
Tail = tail(dataset_tail,y)
print "Tail:"
display(HTML(Tail.to_html()))

print('_'*50)
print '\033[1m'+"Random samples of observations from the data"  
def sample(dataset_sample,z):
    Sample = dataset_sample
    return (Sample)
cString = raw_input("Enter the number, to see random observations: ")
c = int(cString)
z = c
dataset_sample=dataset.sample(z)
Sample = sample(dataset_sample,z)
print "Sample:"
display(Sample)

print('_'*50)
print '\033[1m'+"Choice of observations using slice index"  
def choice(dataset_choice):
    Choice = dataset_choice
    return (Choice)
AAString = raw_input("Enter the starting observations number: ")
AA = int(AAString)
BBString = raw_input("Enter the ending observations number: ")
BB = int(BBString)
dataset_choice=dataset[AA:BB]
Choice = choice(dataset_choice)
print"Choice:"
display(Choice)



print('_'*50)
print '\033[1m'+" Datatype of the data"   
def info(dataset_info):
    Info = dataset_info
    return (Info)

dataset_info=dataset.info()
Info = info(dataset_info)
print "Info:"
display(Info)


print('_'*50)
print '\033[1m'+"Summary statistics of the dataset"    
def describe_f(dataset_describe_f):
    Describe_F = dataset_describe_f
    return (Describe_F)
dataset_describe_f=dataset.describe(include ='all').T
Describe_F = describe_f(dataset_describe_f)
print"Summary Statistics:"
display(Describe_F)



print('_'*50)
# file: Missing_Value_Function.py
def Total_Percent(total_missing,total_cells,missing_data ):  
    percent = ((float(total_missing)/total_cells) * 100)
    missing_data_percent_column = missing_data.sort_values("Percent",ascending= False).head()
    return (percent,missing_data_percent_column)
     
total_cells = np.product(dataset.shape)

dataset_missing_values_sum_per_col = dataset.isnull().sum().sort_values(ascending= False)
columns_with_only_missing_values = dataset_missing_values_sum_per_col.drop(dataset_missing_values_sum_per_col[dataset_missing_values_sum_per_col==0.00].index).sort_values(ascending=False).head()
total_missing = dataset_missing_values_sum_per_col.sum() 


dataset_values_count_per_col = dataset.isnull().count()
dataset_missing_values_percentage_per_col = (dataset_missing_values_sum_per_col/dataset_values_count_per_col)*100
columns_with_only_missing_percent = dataset_missing_values_percentage_per_col.drop(dataset_missing_values_percentage_per_col[dataset_missing_values_percentage_per_col==0.00].index).sort_values(ascending=False).head()


missing_data = pd.concat([columns_with_only_missing_values,columns_with_only_missing_percent],axis=1,keys=['Total', 'Percent'])
percent,missing_data_percent_column= Total_Percent(total_missing,total_cells,missing_data)



print "Total {RED}{}{END} percentage of values are missing in this data set".format(percent,**formatters)
print('_'*50)
if percent !=0:
    print '\033[1m'+ "missing_data_percent_column:"
    display(missing_data_percent_column)
    #Create plot for the columns which have missing values
    fig, axarr = plt.subplots(1, 1, figsize=(15, 8))
    plt.xticks(rotation='90')
    sb.barplot(x=dataset_missing_values_percentage_per_col.index, y=dataset_missing_values_percentage_per_col )
    axarr.set(title= "Percentage of missing data by column", ylabel="Percentage missing" )
else:
    print "{GREEN}Good Luck!!{END} The data has no missing values".format(**formatters)
    
    
```
